The default webcam's frames are continually captured by this code, which shows them in a window called "Laptop camera."
When frame capture fails or you hit the "q" key, it pauses, releases the camera, and shuts down all windows.

```python
import cv2
cap=cv2.VideoCapture(0)
while True:
    ret,frame=cap.read()
    if not ret:
        print("Failed to grab frame")
        break
    cv2.imshow("Laptop camera",frame)

    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

This code outputs each frame's dimensions (height, width, and channels) after reading a video file frame by frame.
Additionally, it extracts and publishes the pixel value in BGR format at location (y=50, x=100) for each frame.

The greyscale movie is shown in a window after this code reads a video file frame by frame and changes each frame from BGR to greyscale.
The loop keeps going until either you hit the "q" key or the video stops, at which point the video capture is released.

```python
import cv2
import numpy as np

cap=cv2.VideoCapture("C:/Users/dsnat/Downloads/videosProject/video/pexels-george-morina-6719160 (2160p).mp4")

while True:
    ret, frame = cap.read()
    if not ret:
        break
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow("video",gray)
    if cv2.waitKey(1)==ord('q'):
        break

cap.release()
```

This code records live video from the camera, shows it in a window called "Shapes," and adds a green rectangle, a blue circle, and the red word "OpenCV Practice" to each frame.
The loop keeps running until you hit "q," at which point the camera is released and all OpenCV windows are closed.

```python
import cv2

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    
    if not ret:
        break

    cv2.rectangle(frame, (100, 100), (300, 300), (0, 255, 0), 3)
    cv2.circle(frame, (200, 200), 50, (255, 0, 0), 3)
    cv2.putText(frame, "OpenCV Practice", (50, 50),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

    cv2.imshow("Shapes", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

This code records live video from the camera, highlights edges in each frame using Canny edge detection, and shows the output in a window called "Edges."
The loop continues until you hit 'q', at which point the camera is released and all OpenCV windows are closed.

```python
import cv2

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    
    if not ret:
        break

    edges = cv2.Canny(frame, 100, 200)
    cv2.imshow("Edges", edges)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

This code uses the camera to record live video and compares successive frames to identify motion.
The motion is shown in a window until you hit "q." It computes the difference between frames, transforms it to greyscale, blurs it to decrease noise, and thresholds it to emphasise moving parts.

```python
import cv2

cap = cv2.VideoCapture(0)

ret, frame1 = cap.read()
ret, frame2 = cap.read()

while True:
    diff = cv2.absdiff(frame1, frame2)

    gray = cv2.cvtColor(diff, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5, 5), 0)

    _, thresh = cv2.threshold(blur, 20, 255, cv2.THRESH_BINARY)

    cv2.imshow("Motion", thresh)
#difference = frame(t) - frame(t+1)
    frame1 = frame2
    ret, frame2 = cap.read()

    if not ret:
        break

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
